```python
def pre_processamento_de_linha(texto):
     
    # Remove tudo que não forem letras
    letters_only_text = re.sub("[^a-zA-Z]", " ", texto)

    # Converte palavras para minúsculo e as separa
    words = letters_only_text.lower().split()

    # Remove stopwords
    set_stopwords = set(stopwords.words("english"))
    meaningful_words = [w for w in words if w not in set_stopwords]
    
    # Stemmiza as palavras
    # ps = PorterStemmer()
    # stemmed_words = [ps.stem(word) for word in meaningful_words]
    
    # Transforma set de palavras em lista
    texto_limpo = " ".join(meaningful_words)

    return texto_limpo
```

```python
df['reviews_sem_acentos'] = [unidecode.unidecode(texto) for texto in  df['review']]
```

```python
df['reviews_pre_processadas'] = df['reviews_sem_acentos'].apply(lambda linha : pre_processamento_de_linha(linha))
```

```python
df.to_csv('imdb_master.csv')
```

```python
reviews_pre_processadas = df['reviews_pre_processadas'] 
```

```python
def tokenizar(sentencas):
    words = []
    for sentenca in sentencas:
        words.extend(sentenca)
        words = sorted(list(set(words)))
    return words
```

```python
#count_vectorizer = TfidfVectorizer(stop_words='english')
count_vectorizer = TfidfVectorizer()
matriz_esparsa = count_vectorizer.fit_transform(reviews_pre_processadas[0:1000])
```

```python
doc_term_matrix = matriz_esparsa.todense()

df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names(), 
                  index= df[0:1000].index)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similaridade =  pd.DataFrame(cosine_similarity(df.loc[[0]], df))
```

```python
similaridade = similaridade.transpose()
```
